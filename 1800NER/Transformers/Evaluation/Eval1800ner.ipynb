```
labels = {
    'LABEL_0': 'O', 
    'LABEL_1': 'B-PRS', 
    'LABEL_2': 'I-PRS', 
    'LABEL_3': 'B-ORG', 
    'LABEL_4': 'I-ORG', 
    'LABEL_5': 'B-LOC', 
    'LABEL_6': 'I-LOC', 
    'LABEL_7': 'B-TME', 
    'LABEL_8': 'I-TME', 
    'LABEL_9': 'B-MSR', 
    'LABEL_10': 'I-MSR', 
    'LABEL_11': 'B-WRK', 
    'LABEL_12': 'I-WRK'
}
```

```
from transformers import pipeline, AutoModelForTokenClassification, AutoTokenizer

model = AutoModelForTokenClassification.from_pretrained("/content/drive/MyDrive/Riksarkivet/1800-talsner/Models/NER1900")
tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/Riksarkivet/1800-talsner/Models/NER1900")

ner_model = pipeline('ner', model=model, tokenizer=tokenizer, device=0)
```

```
numOfDeletedLines = 0
with open('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/Bible2000SentSeg.txt', 'r') as f:
  lines = f.readlines()
with open('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/Bible2000SentSeg.txt', 'w') as f:
  for line in lines:
    tokline = tokenizer(line)
    if len(tokline['input_ids']) <= 512:
      f.write(line)
    else:
      numOfDeletedLines += 1
print(numOfDeletedLines)
```

```
with open('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/valBible.txt', 'r') as f:
  lines = f.readlines()
print(len(lines))
```

```
with open('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/Bible2000SentSeg.txt', 'r') as f:
  linenr = 0
  annotations = []
  for line in f:
    linenr += 1
    f.readline()
    a = ner_model(line)
    for annot in a:
      annot['entity'] = labels[annot['entity']]
    annotations.append(a)
```

```
numOfEntities = 0
for seq in annotations:
  for tok in seq:
    if tok['entity'].startswith('B-LOC'):
      numOfEntities += 1
print(numOfEntities)
```
