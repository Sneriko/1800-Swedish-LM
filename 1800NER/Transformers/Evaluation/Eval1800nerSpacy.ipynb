```
import spacy
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
nlp = spacy.load('/content/drive/MyDrive/Examensarbete/SwedishModel')
```

```
nlp.remove_pipe('ner')
```

```
def runNlp(path):
  with open(path, 'r') as f:
    string = f.readline()
    doc = nlp(string)
    return list(doc.ents)
    
```

```
entitiesForAllParts = []
entitiesForAllParts.append(runNlp('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/bible2000PartI.txt'))
entitiesForAllParts.append(runNlp('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/bible2000PartII.txt'))
entitiesForAllParts.append(runNlp('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/bible2000PartIII.txt'))
entitiesForAllParts.append(runNlp('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/bible2000PartIV.txt'))
entitiesForAllParts.append(runNlp('/content/drive/MyDrive/Riksarkivet/1800-talsner/Datasets/bible2000PartV.txt'))
```

```
numOfEntities = 0
for entitiesForPart in entitiesForAllParts:
  for entity in entitiesForPart:
    if entity.label_ == 'PRS':
      numOfEntities += 1

print(numOfEntities)
```
